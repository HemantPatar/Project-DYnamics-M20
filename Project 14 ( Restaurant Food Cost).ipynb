```python
#Lets seprate the input and output from train dataset
df_x=df_train.drop(columns=["COST"])
y=df_train[["COST"]]
```

```python
#to find random stat which gives maximum r2_score
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
def maxr2_score(regr,df_x,y):
    max_r_score=0
    for r_state in range(42,100):
        x_train, x_test, y_train, y_test = train_test_split(df_x, y,random_state = r_state,test_size=0.20)
        regr.fit(x_train,y_train)
        y_pred = regr.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        print("r2 score corresponding to ",r_state," is ",r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
    print("max r2 score corresponding to ",final_r_state," is ",max_r_score)
    return final_r_state
```

```python
#Lets use decision tree
from sklearn.tree import DecisionTreeRegressor
import numpy as np
dtr=DecisionTreeRegressor()
r_state=maxr2_score(dtr,df_x,y)
```

```python
#lets use random forest regressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import GridSearchCV
import warnings
warnings.filterwarnings("ignore")
rfr=RandomForestRegressor()
parameters = {"n_estimators":[10,100,500]}
clf = GridSearchCV(rfr, parameters, cv=5,scoring="r2")
clf.fit(df_x,y)
clf.best_params_
```

```python
rfr=RandomForestRegressor(n_estimators=500)
r_state=maxr2_score(rfr,df_x,y)
```

```python
import lightgbm as lgb
lg = lgb.LGBMRegressor(silent=False)
param_dist = {"max_depth": [25,50, 75],
              "learning_rate" : [0.01,0.05,0.1],
              "num_leaves": [300,900,1200],
              "n_estimators": [200]
             }
grid_search = GridSearchCV(lg, n_jobs=-1, param_grid=param_dist, cv = 5, scoring="r2")
grid_search.fit(df_x,y)
grid_search.best_estimator_
```

```python
lgbr=lgb.LGBMRegressor(boosting_type='gbdt', class_weight=None, colsample_bytree=1.0,
       importance_type='split', learning_rate=0.05, max_depth=25,
       min_child_samples=20, min_child_weight=0.001, min_split_gain=0.0,
       n_estimators=200, n_jobs=-1, num_leaves=900, objective=None,
       random_state=None, reg_alpha=0.0, reg_lambda=0.0, silent=False,
       subsample=1.0, subsample_for_bin=200000, subsample_freq=0)
r_state=maxr2_score(lgbr,df_x,y)
#Here we see that lightfbm was faster and better than random forest
```

```python
grid_search.best_params_
```

```python
#LGBM also give provision to define which all columns should be treated as categorical
#So we will use above provision and see what is the result
params={'learning_rate': 0.05,'max_depth': 25,'n_estimators': 200,'num_leaves': 900}
cate_features_name=['CITY', 'LOCALITY', 'c1', 'c2', 'c3', 'c4', 'c5','c6', 'c7', 'c8', 't1', 't2']
max_r_score=0
for r_state in range(42,100):
    x_train, x_test, y_train, y_test = train_test_split(df_x, y,random_state = r_state,test_size=0.20)
    d_train = lgb.Dataset(x_train, label=y_train)
    lgb_model = lgb.train(params, d_train, categorical_feature = cate_features_name)
    y_pred = lgb_model.predict(x_test)
    r2_scr=r2_score(y_test,y_pred)
    print("r2 score corresponding to ",r_state," is ",r2_scr)
    if r2_scr>max_r_score:
        max_r_score=r2_scr
        final_r_state=r_state
print("max r2 score corresponding to ",final_r_state," is ",max_r_score)
```

```python
#lets make a function which evaluates the model using cross_val_score
from sklearn.model_selection import cross_val_score
def model_evaluation(model,x,y):
    c_scores=cross_val_score(model,x,y,cv=5,scoring="r2")
    print("Mean r2 score for regressor: ",c_scores.mean())
    print("standard deviation in r2 score for regressor: ",c_scores.std())
    print(c_scores)
```

```python
#The above model i.e light gradient boosting works better when categorical features are defined
#So we will use this to create our final model
#Also when compared to random forest its giving better and fast results
#Lets chcek the cross_val_scores
from sklearn.model_selection import KFold
scores=[]
cate_features_name=['CITY', 'LOCALITY', 'c1', 'c2', 'c3', 'c4', 'c5','c6', 'c7', 'c8', 't1', 't2']
k_fold = KFold(n_splits=5)
for train_index,test_index in k_fold.split(df_x):
    x_train, x_test, y_train, y_test = df_x.iloc[train_index], df_x.iloc[test_index], y.iloc[train_index], y.iloc[test_index]
    d_train = lgb.Dataset(x_train, label=y_train)
    lgb_model = lgb.train(params, d_train, categorical_feature = cate_features_name)
    y_pred = lgb_model.predict(x_test)
    r2_scr=r2_score(y_test,y_pred)
    scores.append(r2_scr)
print("Mean r2 score for lgbm regressor: ",np.array(scores).mean())
print("standard deviation in r2 score for lgbm regressor: ",np.array(scores).std())
print(scores)
```

```python
#lets chcek for random forset
print("RANDOM FOREST\n\n")
model_evaluation(rfr,df_x,y)
```

```python
#From above we see lgbm is performing better so we will use it to make our final predictions
cost_pred=lgb_model.predict(df_test,predict_disable_shape_check=True)
```
